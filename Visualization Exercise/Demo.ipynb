```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
```

```python
dataframe =pd.read_csv("scottish_hills.csv")
dataframe.head(10)
```

```python
sorted_hills=dataframe.sort_values(by=('Height'),ascending=False)
sorted_hills.head(5)
```

```python
x=dataframe.Height
y=dataframe.Latitude
z=dataframe.Longitude
```

```python
plt.scatter(x,y)
plt.savefig("scottish_scatter_plot.png")
```

```python
stats=linregress(x,y)
m=stats.slope
b=stats.intercept
```

```python
plt.scatter(x,y)
plt.plot(x,m*x+b, color='red')
```

```python
#figure size
plt.figure(figsize=(10,10))

#change default marker
plt.scatter(x,y, marker='x')

#set linewidth
plt.plot(x,m*x+b, color='red', linewidth=3)

#add x & y labels set font size
plt.xlabel("Height (m)", fontsize=20)
plt.ylabel("Latitude", fontsize=20)

#Set font size of number labels on the axes
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
```

```python
colors=np.random.rand(len(y)) #gen diff color for diff mountain
plt.scatter(y,z,s=(x-900),c=colors,alpha=0.5)
#s is size, -900 to make difference larger
#c is color
#alpha is transparency
```

### Histogram

```python
#stylesheet
plt.style.use('seaborn-pastel')

#new height variable
shifted_x=x-100
fig,ax=plt.subplots()
ax.hist(x, bins=25, normed=True, histtype='stepfilled', alpha=0.8, label='Height')
ax.hist(shifted_x, bins=25, normed=True, histtype='stepfilled',alpha=0.8, label='Height-100')
ax.legend(prop={'size':10})

ax.set_ylabel('Normalised distribution')
ax.set_xlabel('Height')
```

### Bar Charts

```python
labels=['EU','US','AS']

meateaters=[122,135,80]
vegetarians=[40,43,23]

#keep numeric idex of x-axis labels. position of ticks
index=np.arange(len(labels))

fig,ax=plt.subplots()

bar_width=0.35
opacity=0.6

rects1=ax.bar(index, meateaters, bar_width,
              alpha=opacity,
              color='b',
              label='Meat eaters')
rects2=ax.bar(index+bar_width, vegetarians, bar_width,
              alpha=opacity,
              color='r',
              label='Veges')

ax.set_xlabel('Continent')
ax.set_ylabel('c02 emission')
ax.set_title('c02 emission per continent, per diet')

ax.set_xticks(index+bar_width/2)

ax.set_xticklabels(labels)
ax.legend()

```

### Line Plots

```python
def fheit2celsius(temp):
    return (5./9.)*(temp-32)
    
temperature=[100,102,106,105,90,85,85,89,100,102,103,108,102,106,105,90,85,85,89,100,102,103,108]
```

```python
fig, ax_f=plt.subplots()
ax_c=ax_f.twinx() #Create a twin Axes sharing the xaxis

#plot data
ax_f.plot(temperature)
ax_f.set_xlim(0,24) #24hours

#axis limits
y1,y2=ax_f.get_ylim()
ax_c.set_ylim(fheit2celsius(y1),fheit2celsius(y2))
ax_c.figure.canvas.draw()

#change axis labels
ax_f.set_title('2scales: Fheight and Celsius')
ax_f.set_ylabel('Fheight')
ax_c.set_ylabel('celsius')
ax_f.set_xlabel('Time (hour of day)')
```
